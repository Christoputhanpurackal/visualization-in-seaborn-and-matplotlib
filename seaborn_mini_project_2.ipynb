```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/user_behavior_dataset.csv')
data
```

```python
sns.set(style='whitegrid')
```

```python
# Plot the data
plt.figure(figsize=(12, 6))
sns.lineplot(data, x='Device Model', y='Operating System', marker='o')

plt.title("App Usage Time (min/day) by User ID")
plt.xlabel("User ID")
plt.ylabel("App Usage Time (min/day)")
plt.grid(True)
plt.show()
```

```python
# Create the stem plot using Matplotlib
plt.figure(figsize=(12, 6))
(markerline, stemlines, baseline) = plt.stem(data['Device Model'], data['Number of Apps Installed'], basefmt=" ")

# Customize the plot using Seaborn
sns.set(style="whitegrid")
plt.setp(markerline, 'markerfacecolor', 'b')
plt.setp(stemlines, 'linestyle', '-', 'color', 'r', 'linewidth', 1.5)
plt.setp(baseline, 'color', 'k', 'linewidth', 1)
plt.title('Device Model vs Number of Apps Installed')
plt.xlabel('Device Model')
plt.ylabel('Number of Apps Installed')
plt.grid(True)
plt.show()
```

```python
# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data,x='Device Model', y='Number of Apps Installed',palette='viridis',hue='Gender')
plt.xticks(rotation=90)
plt.title('Device Model vs Number of Apps Installed')
plt.xlabel('Device Model')
plt.ylabel('Number of Apps Installed')
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(12, 6))
sns.histplot(data['App Usage Time (min/day)'], bins=10,edgecolor='black',kde='True')

plt.title("Distribution of App Usage Time (min/day)")
plt.xlabel("App Usage Time (min/day)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data, x='App Usage Time (min/day)', y='Battery Drain (mAh/day)',hue='Gender', style='Operating System')

plt.title("Scatter Plot of App Usage Time vs Battery Drain")
plt.xlabel("App Usage Time (min/day)")
plt.ylabel("Battery Drain (mAh/day)")
plt.legend(title='Gender and OS', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
user_ids = data['User ID']
app_usage_time = data['App Usage Time (min/day)']
screen_on_time = data['Screen On Time (hours/day)']
battery_drain = data['Battery Drain (mAh/day)']
# Plot the stack plot
plt.figure(figsize=(12, 6))
plt.stackplot(user_ids, app_usage_time, screen_on_time, battery_drain, labels=['App Usage Time', 'Screen On Time', 'Battery Drain'], colors=['skyblue', 'lightgreen', 'lightcoral'])

plt.title("Stack Plot of App Usage Time, Screen On Time, and Battery Drain by User ID")
plt.xlabel("User ID")
plt.ylabel("Values")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Create a boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data,x='Gender', y='App Usage Time (min/day)', palette='pastel', showfliers=True,
    notch=True )

plt.title("Boxplot of App Usage Time (min/day) by Gender")
plt.xlabel("Gender")
plt.ylabel("App Usage Time (min/day)")
plt.show()
```

```python
# Count the number of occurrences for each gender
gender_counts = data['Device Model'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))

plt.title("Device Distribution")
plt.show()
```

```python
# Calculate the correlation matrix
corr = data[['App Usage Time (min/day)', 'Screen On Time (hours/day)', 'Battery Drain (mAh/day)', 'Number of Apps Installed', 'Data Usage (MB/day)', 'Age', 'User Behavior Class']].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=.5)

plt.title("Heatmap of Correlations")
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
# Create a countplot for Gender
plt.figure(figsize=(10, 6))
sns.countplot(data, x='Gender', palette='pastel')

plt.title("Count of Genders")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

```

```python
# Create a pairplot
plt.figure(figsize=(10, 10))
sns.pairplot(data, vars=['App Usage Time (min/day)', 'Screen On Time (hours/day)', 'Battery Drain (mAh/day)', 'Number of Apps Installed', 'Data Usage (MB/day)', 'Age'], hue='Gender', palette='pastel')

plt.show()
```

```python

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: App Usage Time by User ID
sns.lineplot(data,ax=axes[0, 0], x='User ID', y='App Usage Time (min/day)', marker='o')
axes[0, 0].set_title('App Usage Time (min/day) by User ID')

# Plot 2: Screen On Time by User ID
sns.lineplot(data,ax=axes[0, 1], x='User ID', y='Screen On Time (hours/day)', marker='o')
axes[0, 1].set_title('Screen On Time (hours/day) by User ID')

# Plot 3: Battery Drain by User ID
sns.lineplot(data,ax=axes[1, 0], x='User ID', y='Battery Drain (mAh/day)', marker='o')
axes[1, 0].set_title('Battery Drain (mAh/day) by User ID')

# Plot 4: Data Usage by User ID
sns.lineplot(data,ax=axes[1, 1], x='User ID', y='Data Usage (MB/day)', marker='o')
axes[1, 1].set_title('Data Usage (MB/day) by User ID')

# Adjust layout
plt.tight_layout()
plt.show()
```
