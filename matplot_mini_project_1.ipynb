```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/user_behavior_dataset.csv')
data.head()
```

```python
#  Plot the data
plt.figure(figsize=(12, 6))
plt.plot(data['Operating System'], data['Device Model'], marker='o', linestyle='-', color='b')
plt.title('Operating System vs Device Model')
plt.xlabel('Operating System')
plt.ylabel('Device Model')
plt.grid(True)
plt.show()


```

```python
# stem plot Device Model vs Number of Apps Installed
plt.figure(figsize=(7,10))
plt.stem(data['Device Model'],data['Number of Apps Installed'])
plt.title('Device Model vs Number of Apps Installed')
plt.xlabel('Device Model')
plt.ylabel('Number of Apps Installed')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.bar(data['Device Model'],data['Number of Apps Installed'],color='lightblue')
plt.title('Device Model vs Number of Apps Installed')
plt.xlabel('Device Model')
plt.ylabel('Number of Apps Installed')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.hist(data['App Usage Time (min/day)'],bins=10,edgecolor='black')
plt.title('App Usage Time (min/day)')
plt.xlabel('App Usage Time (min/day)')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

```python
# scatter plot of data usage and battery drain
plt.figure(figsize=(10, 5))
plt.scatter(data['Data Usage (MB/day)'],data['Battery Drain (mAh/day)'] ,alpha=0.6, edgecolors='w')
plt.title('Data Usage (MB/day) vs Battery Drain (mAh/day)')
plt.xlabel('Data Usage (MB/day)')
plt.ylabel('Battery Drain (mAh/day)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.stackplot(data.index,data['Age'],  data['Screen On Time (hours/day)'],data['Number of Apps Installed'],
               labels=['AGE', 'Screen time', 'apps'], alpha=0.8)

plt.title('User Behavior Over Time')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.boxplot()
plt.title('Box Plot of User Behavior Metrics')
plt.ylabel('Hours / Percentage')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Count the number of occurrences for each gender
gender_counts = data['Gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])

plt.title("Gender Distribution")
plt.show()

```
